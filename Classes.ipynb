# Patient CLASS

```python
class Patient():
    def __init__(self, first_name,last_name,age,gender,date,bill):
        self.first_name = first_name
        self.last_name  = last_name
        self.age        = age
        self.gender     = gender
        self.date       = date
        self.bill       = bill
        #pass
    def minor(self):
        if self.age < 21:
            print(self.first_name +' '+"is a minor")
        else:
            print(self.first_name +' '+"is an adult")
    def checkup(self):
        print("Patient Checkup date was"+' '+self.date)
    def billing(self):
        print("Patient Bill was"+' '+str(self.bill))
```

```python
pid01 = Patient("Taleb","Siddique",43,"M",'01/03/2019',500)
pid02 = Patient("Sheikh","Zaid",34,"M",'04/02/2019',300)
pid03 = Patient("Maria","Saleem",23,"F",'07/04/2019',800)
pid04 = Patient("Sara","Siddique",45,"F",'12/04/2019',1000)
pid05 = Patient("Haroon","Chaiwala",20,"M",'10/04/2019',200)
```

```python
pid05.minor()
pid05.checkup()
pid05.billing()
```

```python
pid03.billing()
pid03.billing()
pid03.checkup()
```

# DOG CLASS

```python
class Dog():
    def __init__(self, name,breed,color, age,gender,height,weight,speed):
        self.name  = name
        self.breed = breed
        self.color = color
        self.age   = age
        self.gender = gender
        self.height = height
        self.weight = weight
        self.speed  = speed
    def height_func(self):
        print("The dog breed is {0} & the average height of {2} are {1} cm".format(self.breed,self.height,self.gender))
    def weight_func(self):
        print("The dog breed is {0} & the average weight of {2} are {1} kg".format(self.breed,self.weight,self.gender))
    def speed_func(self):
        print("The dog breed is {0} & the average speed of {2} are {1} mph".format(self.breed,self.speed,self.gender))   
```

```python
dober1 = Dog('Dober','Dobermann','Black',4,'Male','66–72','40–45',35)
dober2 = Dog('Dober','Dobermann','Black',3,'Female','61–68','32–35',33)
dan3   = Dog('Great Dan','Great Dan','Black',4,'Male','76–86','54–90',30)
dan4   = Dog('Great Dan','Great Dan','Black',4,'Female','71–81','45–59',30)
Grey5  = Dog('Greyhound','Greyhound','brown',4,'Male','71–76','27–40',45)
```

# Restaurant

```python
class Restaurant():
    def __init__(self,restaurant_name,cuisine_type):
        self.restaurant_name = restaurant_name
        self.cuisine_type    = cuisine_type
    def describe_restaurant(self):
        print("The restaurant Name is {0} and Cuisine type is {1} ".format(self.restaurant_name,self.cuisine_type))
    def open_restaurant(self):
        print("{0} is Open till 12 o clock".format(self.restaurant_name))
    
```

```python
res1 = Restaurant("Kababjees","Desi")
res2 = Restaurant("Charcoal","Fast food")
res3 = Restaurant("Broadway","Pizza")
```

```python
res3.describe_restaurant()
res3.open_restaurant()
```

```python
class User():
    def __init__(self,first_name,last_name):
        self.first_name = first_name
        self.last_name  = last_name
    def describe_user(self):
        print("{0} & {1} ".format(self.first_name,self.last_name))
    def greet_user(self):
        print("Hello, {0} {1} Welcome in our Restaurant ".format(self.first_name,self.last_name) )
```

```python
usr01 = User("Kamran","saleem")
```

```python
usr01.greet_user()
```
